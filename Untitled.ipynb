```python
!pip install streamlit
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
st.text(report)
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import streamlit as st

def detect_anomalies(log_df):
    model = IsolationForest(contamination=0.1)
    model.fit(log_df.select_dtypes(include=[np.number]))
    log_df['anomaly'] = model.predict(log_df.select_dtypes(include=[np.number]))
    return log_df[log_df['anomaly'] == -1]

def classify_threat(entry):
    keywords = {
        'phishing': ['login', 'password', 'credential'],
        'ddos': ['traffic', 'requests'],
        'malware': ['download', 'file', 'infected']
    }
    entry = entry.lower()
    for threat, words in keywords.items():
        if any(word in entry for word in words):
            return threat.capitalize()
    return 'Normal'

def build_report(df):
    counts = df['threat_class'].value_counts()
    report = "Summary Report:\n"
    for threat, count in counts.items():
        if threat != 'Normal':
            report += f"- {threat}: {count} incident(s)\n"
    return report if counts.shape[0] > 0 else "No significant threats detected."

def run_autoshield_pipeline():
    data = {
        "cpu": [50, 99, 45, 87, 95, 20],
        "network_io": [100, 2000, 120, 1800, 1600, 80],
        "log_entry": [
            "User login success",
            "Multiple failed logins from unknown IP",
            "System update installed",
            "High outbound traffic from server",
            "Multiple requests to admin panel",
            "File downloaded"
        ]
    }
    df = pd.DataFrame(data)
    df['threat_class'] = df['log_entry'].apply(classify_threat)
    anomalies = detect_anomalies(df)
    report = build_report(df)
    return df, anomalies, report

st.title("AutoShield: Autonomous Cyber Threat Hunter")
df, anomalies, report = run_autoshield_pipeline()

st.subheader("Analyzed Logs")
st.dataframe(df)

st.subheader("Detected Anomalies")
st.dataframe(anomalies)

st.subheader("Threat Summary Report")
st.text(report)
```
